```python
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# prompt: read tea_data_corrected.csv in pandas
df = pd.read_csv('tea_data_corrected.csv')
df
```

```python
# Set display options for floating-point numbers
pd.set_option('display.float_format', '{:.7f}'.format)
```

```python
df["tea_dum"] = (df["teasown"] > 0).astype(float)
df["post"] = (df["biryr"] >= 1979).astype(float)

print(f'tea_dum mean: {df["tea_dum"].mean()} and post mean: {df["post"].mean()}')
```

```python
formula = 'sex ~ tea_dum + post + tea_dum*post'

# Fit the model using formula API
model = smf.ols(formula=formula, data=df).fit()

# Print the summary
print(model.summary())
```

```python
model_df = pd.DataFrame({
    'Coefficient': model.params,
    'Std. Error': model.bse,
    't-value': model.tvalues,
    'P>|t|': model.pvalues
})

model_df
```

```python
df["orchard_dum"] = (df["orch"] > 0).astype(float)
df["cashcrop_dum"] = (df["cashcrop"] > 0).astype(float)
```

```python
formula = 'sex ~ tea_dum + orchard_dum + cashcrop_dum + post + tea_dum*post + orchard_dum*post + cashcrop_dum*post'

# Fit the model using formula API
model = smf.ols(formula=formula, data=df).fit()

# Print the summary
print(model.summary())
```

```python
model_df = pd.DataFrame({
    'Coefficient': model.params,
    'Std. Error': model.bse,
    't-value': model.tvalues,
    'P>|t|': model.pvalues
})

model_df
```

```python
formula = 'sex ~ tea_dum*post + orchard_dum*post + cashcrop_dum*post + C(admin) + C(biryr)'

# Fit the model using formula API
model = smf.ols(formula=formula, data=df).fit()

# Print the summary
print(model.summary())
```

```python
model_df = pd.DataFrame({
    'Coefficient': model.params,
    'Std. Error': model.bse,
    't-value': model.tvalues,
    'P>|t|': model.pvalues
})

model_df
```
